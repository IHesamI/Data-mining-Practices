```python
import os
import calendar
import numpy as np
import networkx as nx
import pandas as pd
from pandas.plotting import scatter_matrix, parallel_coordinates
import seaborn as sns
from sklearn import preprocessing
import matplotlib.pyplot as plt
import datetime as dt
```

```python
Boston_df=pd.read_csv('BostonHousing.csv')
    # ^ TWO WAYS OF CHANGING THE NAME OF THE COLUMNS
    # * Boston_df.columns=[i.replace('. ','_') for i in Boston_df.columns]
    # * Boston_df=Boston_df.rename(columns={'CAT. MEDV':'CAT_MEDV'})
Boston_df=Boston_df.rename(columns={'CAT. MEDV':'CAT_MEDV'})
Boston_df.plot(kind='scatter',x='LSTAT',y='MEDV',legend=False)
```

```python
ax=Boston_df.groupby('CHAS').mean().MEDV.plot(kind='bar')
ax.set_ylabel('% CAT_MEDV')
```

```python
dataforplot=Boston_df.groupby('CHAS').mean()['CAT_MEDV']*100
ax=dataforplot.plot(kind='bar',figsize=[5,3])
ax.set_ylabel('% CAT_MEDV')
```

```python
Amtrak_df=pd.read_csv('Amtrak.csv',squeeze=True)
Amtrak_df['Date']=pd.to_datetime(Amtrak_df.Month,format='%d/%m/%Y')
ridership_ts=pd.Series(Amtrak_df.Ridership.values,index=Amtrak_df.Date)
ridership_ts.plot(ylim=[1300,2300],legend=False)
plt.xlabel('Year')
plt.ylabel('Ridership')
```
